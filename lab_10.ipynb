###Implement K-means clust

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import numpy as np
```

```python
from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50,color='blue');
```

Import K_Means from SKlearn and Fit the data

```python
list(X.shape)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, n_init=10)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

Visualize the fitted data by coloring the blobs by assigned list item list item label numbers

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers=kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1],c='black', s=200, alpha=0.5);
```

How many number of Clusters

plot the clusters formed using scatter plot

```python
labels = KMeans(10, random_state=0, n_init=10).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels,  s=50, cmap='viridis');
```
